```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("datos_aerodinamicos.csv")
```

```python
print(data)
```

```python
plt.plot(data["velocidad"], data["drag"])
plt.xlabel("Velocidad (m/s)")
plt.ylabel("Drag")
plt.title("Relación Velocidad vs Drag")
plt.savefig("grafico_drag.png")
plt.show()
```
